# Instagram Private API

```python
# imports
from instagrapi import Client
```

```python
# account keys
ACCOUNT_USERNAME = 'ringledingleshingle'
ACCOUNT_PASSWORD = 'nobatteriesinvolved!'
```

```python
# account login
cl = Client()
cl.login(ACCOUNT_USERNAME, ACCOUNT_PASSWORD)
print('login complete')
```

```python
# get personal account info
cl.account_info().dict()
# can also change things like bio, name, etc. with cl.account_edit()
# docs: https://adw0rd.github.io/instagrapi/usage-guide/account.html
```

```python
# get user id based on username
TARGET = 'robertpatel21'#'hannahta0' #
user_id = cl.user_id_from_username(TARGET)
print(user_id)
#medias = cl.user_medias(cl.user_id, 20)
```

```python
# get acc media
medias = cl.user_medias(user_id, 20)
print(medias)
# write medias to a file
with open('medias.txt', 'w') as f:
    f.write(str(medias))
f.close()
```

```python
# convert medias entries to dict entries
medias_dict = [medias[i].dict() for i in range(len(medias))]
print(medias_dict)
```

```python
# what keys are avaialable
print(medias_dict[0].keys())
```
